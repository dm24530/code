# 模型构建

```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
inputfile = 'zscoreddata.xls'
```

需要进行的聚类类别数

```python
k = 5
```

```python
data = pd.read_excel(inputfile)
data = data[['ZR','ZF','ZM','ZC','ZL']]
data
```

调用k-means算法，进行聚类分析

```python
kmodel = KMeans(n_clusters = k, n_jobs = 4)
kmodel.fit(data)
```

查看聚类中心

```python
kmodel.cluster_centers_
```

```python
kmodel.labels_
```

```python
from pandas import DataFrame,Series
```

```python
r1 = DataFrame(kmodel.cluster_centers_)
r1
```

```python
type(r1)
```

```python
r1.iloc[0]
```

查看各样本对应的类别

```python
kmodel.labels_
```

```python
r2 = Series(kmodel.labels_)
r2
```

```python
r3 = pd.concat([r1, r2], axis=1)
r3
```

# 绘制雷达图

```python
import pyecharts.options as opts
from pyecharts.charts import Radar
import numpy as np
```

```python
v1 = [np.array(r1.iloc[4]).tolist()]
v2 = [np.array(r1.iloc[0]).tolist()]
v3 = [np.array(r1.iloc[1]).tolist()]
v4 = [np.array(r1.iloc[2]).tolist()]
v5 = [np.array(r1.iloc[3]).tolist()]
```

```python
r1
```

```python
r1.iloc[0]
```

```python
(
    Radar(init_opts=opts.InitOpts(width="1280px", height="720px", bg_color="#CCCCCC"))
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name="ZR", max_=300),
            opts.RadarIndicatorItem(name="ZF", max_=100),
            opts.RadarIndicatorItem(name="ZM", max_=200000),
            opts.RadarIndicatorItem(name="ZC", min_=-5),
            opts.RadarIndicatorItem(name="ZL", max_=100),
        ],
        splitarea_opt = opts.SplitAreaOpts(
            is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
        ),
        textstyle_opts = opts.TextStyleOpts(color="#fff"),
    )
    .add(
        series_name="客户群1",
        data = v1,
        linestyle_opts = opts.LineStyleOpts(color="#5D0000")
    )
     .add(
        series_name="客户群2",
        data = v2,
        linestyle_opts = opts.LineStyleOpts(color="#CD0454")
    )
     .add(
        series_name="客户群3",
        data = v3,
        linestyle_opts = opts.LineStyleOpts(color="#6D4521")
    )
     .add(
        series_name="客户群4",
        data = v4,
        linestyle_opts = opts.LineStyleOpts(color="#999999")
    )
     .add(
        series_name="客户群5",
        data = v5,
        linestyle_opts = opts.LineStyleOpts(color="#FD3542")
    )
    .set_series_opts(label_opts = opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts = opts.TitleOpts(title="基础雷达图"), legend_opts = opts.LegendOpts()
    )
    .render("chart.html")
)
```
