```python
#CrawUnivRankingA.py
import requests
from bs4 import BeautifulSoup
import bs4
 
def getHTMLText(url):
    try:
        r = requests.get(url, timeout=30)
        r.raise_for_status()
        r.encoding = "GBK"
        return r.text
    except:
        return "获取失败"
 

def fillDicList(ulist,html):
    soup = BeautifulSoup(html,'html.parser')
#     for tr in soup.body.children:
#         print(tr)
#         for td in soup.a:
#             print(td.string)
    
#     for tr1 in soup.find_all('span'):
#         print(tr1.string)
        
    for tr in soup.find_all('a','fontbox2'):
#         print(tr.get_text())
        tds = tr.get_text()
        ulist.append([tds[0],tds[1:-1]])
#         print(ulist)
        

def printDicList(ulist, num):
    print("{:^10}\t{:^10}".format("汉字","拼音"))
    for i in range(num):
        u=ulist[i]
        print("{:^12}\t{:^12}".format(u[0],u[1]))
     
def main():
    uinfo = []
    depth = 4
    start_url = 'http://xh.5156edu.com/wx/'
    infoList = []
    for i in range(1,depth):
        try:
            if i == 1:
                url = start_url + 'tu' + '.html'
            else:
                url = start_url + 'tu_' + str(i) + '.html'
                html = getHTMLText(url)
#             print(url)
        except:
            return 'main错误'
        html = getHTMLText(url)
        fillDicList(uinfo, html)
    length = len(uinfo)
    printDicList(uinfo, length) 
        
        
main()
```
