## 最简单的折线图

## 折线图
绘制多条曲线、曲线颜色、线型、标记等参数设置


```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

yy=[1,2,3,4,5,3,1,2,7,8] #随便创建的数据
xx=[3,5,4,1,9,3,2,5,6,3]
zz=[2,2,4,7,4,8,2,4,5,6]

plt.plot(xx, color='r', linestyle='-', label ="Data 1")
plt.plot(yy, color='g', linestyle='--', label ="Data 2")
plt.plot(zz, color='b', linestyle=':', label ="Data 3")
plt.legend()
plt.xlabel("x轴名称", fontproperties="simhei")
plt.ylabel("y轴名称", fontproperties="simhei")
plt.title("折线图进阶", fontproperties="simhei")

plt.show()
```

## 散点图实战

#### 绘制普通散点图：10个点位置随机的散点图

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x = np.random.rand(N)
y = np.random.rand(N)
plt.scatter(x, y)
plt.show()
```

#### 改变散点大小：每个点随机大小

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x = np.random.rand(N)
y = np.random.rand(N)

size = (30 *np.random.rand(N))**2

plt.scatter(x, y, s =size)
plt.show()
```

#### 更改散点颜色、透明度：颜色随机，透明度设为0.5

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x = np.random.rand(N)
y = np.random.rand(N)

size = (30 *np.random.rand(N))**2

color = np.random.rand(N)

plt.scatter(x, y, s =size, c=color, alpha = 0.5)
plt.show()
```

#### 更改散点形状：点形状改为上三角

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x = np.random.rand(N)
y = np.random.rand(N)

size = (30 *np.random.rand(N))**2

color = np.random.rand(N)

plt.scatter(x, y, s =size, c=color, alpha = 0.5, marker='^')
plt.show()
```

#### 一张图绘制两组数据的散点图

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x1 = np.random.rand(N)
y1 = np.random.rand(N)

x2 = np.random.rand(N)
y2 = np.random.rand(N)

plt.scatter(x1, y1,  alpha = 0.5, marker='^')
plt.scatter(x2, y2, alpha = 0.5, marker='o')
plt.show()
```

#### 为散点图增加图例

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x1 = np.random.rand(N)
y1 = np.random.rand(N)

x2 = np.random.rand(N)
y2 = np.random.rand(N)

plt.scatter(x1, y1,  alpha = 0.5, marker='^', label='triangle')
plt.scatter(x2, y2, alpha = 0.5, marker='o', label = 'circle')

plt.legend(loc='best')

plt.show()
```
