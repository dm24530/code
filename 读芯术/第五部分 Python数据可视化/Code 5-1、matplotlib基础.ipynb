## Matplotlib Gallery示例

```python
import numpy as np
import matplotlib.pyplot as plt
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars
p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

## 快速绘图示例


#### 绘制sin(x),和cos(x**2)两条曲线，要求x取值范围0至10、x轴设为"Time(s)"、y轴设为"Volt"、图名称为“PyPlot First Example”，并显示图例。

##### 步骤：1、导入numpy和matplotlib库；2、计算绘图数据；3、绘制曲线、4、设置相关参数

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x**2)

plt.plot(x, y, "r-", label="$sin(x)$",linewidth=2)
plt.plot(x, z, "b--",label="$cos(x^2)$")

plt.xlabel("Times(s)")
plt.ylabel("Volt")
plt.title("PyPlot First Example")
plt.legend()

plt.show()
```

## 子图绘制subplot

#### 上下两个子图，直接使用matplotlib

```python
import matplotlib.pyplot as plt

plt.subplot(2,1,1)
plt.xticks([]), plt.yticks([])
plt.text(0.5,0.5, 'subplot(2,1,1)',ha='center',va='center',size=24,alpha=.5)

plt.subplot(2,1,2)
plt.xticks([]), plt.yticks([])
plt.text(0.5,0.5, 'subplot(2,1,2)',ha='center',va='center',size=24,alpha=.5)

# plt.savefig('../figures/subplot-horizontal.png', dpi=64)
plt.show()
```

matplotlib通过pyplot模块提供了一套和MATLAB类似的绘图API,matplotlib还提供了一个名为pylab的模块，其中包括了许多NumPy和pyplot模块中常用的函数，方便用户快速进行计算和绘图，十分适合在IPython交互式环境中使用, 但两者有什么区别呢？

pylab结合了pyplot和numpy，对交互式使用来说比较方便，既可以画图又可以进行简单的计算。但是，对于一个项目来说，建议分别倒入使用，即：

import numpy as np

import matplotlib.pyplot as plt

而不是
import pylab as pl
或
from pylab import *

#### 左右两个子图， 使用pylab

```python
from pylab import *

subplot(1,2,1)
xticks([]), yticks([])
text(0.5,0.5, 'subplot(1,2,1)',ha='center',va='center',size=24,alpha=.5)

subplot(1,2,2)
xticks([]), yticks([])
text(0.5,0.5, 'subplot(1,2,2)',ha='center',va='center',size=24,alpha=.5)

show()
```
