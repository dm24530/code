# 阶跃函数

```python
from numpy import array, random, dot
import numpy as np
```

```python
x = array([[1,-0.5,-0.5],
      [1,-0.5,0.5],
      [1,0.3,-0.5],
      [1,0,1]])
y = array([[1.0],
      [1.0],
      [0],
      [0]])
```

```python
def step(x): 
  for i in range(len(x)):
    if x[i] > 0:
      x[i] = 1
      # print(x[i])
    else:
      x[i] = 0
  return x
```

```python
weights = np.random.random([3,1])
weights
```

```python
d = 0.01
judge = 0
```

```python
def update():
  global weights, judge, x, y
  judge = step(np.dot(x, weights))
  weights2 = d * (x.T.dot(y-judge))/int(x.shape[0])
  weights += weights2
```

```python
for i in range(1000):
  update()
  judge = step(np.dot(x, weights))
  if(judge == y).all():
    break
print(judge)
print(i)
```

# 单层感知器（sgn）

```python
from numpy import array, dot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 输入数据
X = np.array([[1,3,3],
        [1,4,3],
        [1,1,1],
        [1,0,2]])
# 输入标签
Y = np.array([[1],
        [1],
        [-1],
        [-1]])
```

```python
W = (np.random.random([3,1]) - 0.5) * 2
W
```

学习率设置

```python
lr = 0.01
```

神经网络输出

```python
O = 0
```

更新权重


image.png

```python
O = np.sign(np.dot(X, W))
O
```

```python
X.T.dot(Y-O)/int(X.shape[0])
```

```python
def update():
  global X,Y,W,lr
  O = np.sign(np.dot(X, W))
  W_C = lr * (X.T.dot(Y-O))/int(X.shape[0])
  W += W_C
```

```python
O
```

```python
for i in range(100):
  update()
  print(W)
  print(O)
  O = np.sign(np.dot(X, W))
  if(O == Y).all():
    print("Finished")
    print("echo:", i)
    break
```
