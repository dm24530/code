```python
# 用列表存储数据，并计算这组数据的平方
a = [1,2,3,4]
c = []
for i in a:
    c.append(i**2)
print(c)

# 用ndarray类型来存储，计算显然更加加单
# ndarray具有广播功能
import numpy as np
a = np.array([1,2,3,4])
a**2
```

```python
# Python原生数据类型与numpy数组性能对比
def add(n):
    a = []
    b = []
    c = []
    for i in range(n):
        a.append(i ** 2)
        b.append(i + 3)
        c.append(a[i] + b[i])
    return c

def np_add(n):
    a = np.arange(n) ** 2
    b = np.arange(n) + 3
    c = a + b
    return c

N = 1000000
# 比较性能，使用“魔法函数” timeit
%timeit add(N)
%timeit np_add(N)
# 使用numpy运行时间是使用列表的70分之一
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/country_data_1996_2015.csv', index_col='时间')
df.fillna(value=0, inplace=True)
df
```

```python
# 计算2015年各个国家人口数据

data_2015 = [list(z) for z in zip(np.array(df.columns), np.array(df.loc[2015], dtype='int'))]
data_2015
```

```python
# 计算朝鲜历史各个时期人口数据

data_cx = df['朝鲜']
data_cx
```

```python
# 计算缅甸2014年的人口数据
data_md_2014 = ['缅甸', df.loc[2014]['缅甸']]
data_md_2014
```

```python
# 计算每一个国家历史平均人口数据
data_mean = np.array(np.mean(df, axis=0), dtype='int')
[list(z) for z in zip(df.columns, data_mean)]
```

```python
# 计算亚洲2015年总人口，及平均人口
print('亚洲2015年总人口: {:.0f}'.format(np.array(df.loc[2015]).sum()))
```

```python
# 计算印度、柬埔寨、阿富汗在2011、2012、2013年总人口及平均人口
data_yja = df[['印度', '柬埔寨', '阿富汗']]
data_yja_list = data_yja.loc[[2011, 2012, 2013]]
count = data_yja_list.sum().tolist()
mean = [int(x) for x in data_yja_list.mean().tolist()]
print('印度、柬埔寨、阿富汗在2011、2012、2013年总人口: \n印度: {:.0f}, 柬埔寨: {:.0f}, 阿富汗: {:.0f}'.format(count[0], count[1], count[2]))
print('平均人口: \n印度: {:.0f}, 柬埔寨: {:.0f}, 阿富汗: {:.0f}'.format(mean[0], mean[1], mean[2]))
```

```python
# 计算任意两个国家之间的人口差数据
print('印度和柬埔寨1996年-2015年时期的人口差距')
[list(z) for z in zip(df.index, np.array(df['印度']) - np.array(df['柬埔寨']))]
```

```python
from PIL import Image
import matplotlib.pyplot as plt
image = Image.open('images/timg.jfif')
image = np.array(image)
image.shape
```

```python
%matplotlib inline
plt.imshow(image)
```

```python
# 垂直方向翻转
# 这里使用数组切片的方式来完成，
# 相当于将图片最后一行挪到第一行，
# 倒数第二行挪到第二行，..., 
# 第一行挪到倒数第一行
# 对于行指标，使用::-1来表示切片，
# 负数步长表示以最后一个元素为起点，向左走寻找下一个点
# 对于列指标和RGB通道，仅使用:表示该维度不改变
image2 = image[::-1, :, :]
plt.imshow(image2)
```

```python
image3 = image[:, ::-1, :]
plt.imshow(image3)
```

```python
image4 = image[::-1, ::-1, :]
plt.imshow(image4)
```

```python
im = Image.fromarray(image4)
im.save('images/im.jpg')
plt.imshow(Image.open('images/im.jpg'))
```

```python
H, W = image.shape[0], image.shape[1]
H1 = H // 2
H2 = H
image5 = image[H1:H2, :, :]
plt.imshow(image5)
```

```python
W1 = W // 6
W2 = (W // 3) * 2
image6 = image[:, W1:W2, :]
plt.imshow(image6)
```

```python
image7 = image[H1:H2, W1:W2, :]
plt.imshow(image7)
```

```python
image8 = image * 0.5
plt.imshow(image8.astype('uint8'))
```

```python
# 调整亮度
image9 = image * 2.0
# 由于图片的RGB像素值必须在0-255之间，
# 此处使用np.clip进行数值裁剪
image9 = np.clip(image9, \
        a_min=None, a_max=255.)
plt.imshow(image9.astype('uint8'))
```

```python
#高度方向每隔一行取像素点
image10 = image[::2, :, :]
plt.imshow(image10)
```

```python
#宽度方向每隔一列取像素点
image11 = image[:, ::2, :]
plt.imshow(image11)
```

```python
#间隔行列采样，图像尺寸会减半，清晰度变差
image12 = image[::2, ::2, :]
plt.imshow(image12)
image10.shape
```
